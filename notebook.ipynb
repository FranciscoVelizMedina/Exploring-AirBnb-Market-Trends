Welcome to New York City, one of the most-visited cities in the world. There are many Airbnb listings in New York City to meet the high demand for temporary lodging for travelers, which can be anywhere between a few nights to many months. In this project, we will take a closer look at the New York Airbnb market by combining data from multiple file types like `.csv`, `.tsv`, and `.xlsx`.

Recall that **CSV**, **TSV**, and **Excel** files are three common formats for storing data. 
Three files containing data on 2019 Airbnb listings are available to you:

**data/airbnb_price.csv**
This is a CSV file containing data on Airbnb listing prices and locations.
- **`listing_id`**: unique identifier of listing
- **`price`**: nightly listing price in USD
- **`nbhood_full`**: name of borough and neighborhood where listing is located

**data/airbnb_room_type.xlsx**
This is an Excel file containing data on Airbnb listing descriptions and room types.
- **`listing_id`**: unique identifier of listing
- **`description`**: listing description
- **`room_type`**: Airbnb has three types of rooms: shared rooms, private rooms, and entire homes/apartments

**data/airbnb_last_review.tsv**
This is a TSV file containing data on Airbnb host names and review dates.
- **`listing_id`**: unique identifier of listing
- **`host_name`**: name of listing host
- **`last_review`**: date when the listing was last reviewed

```python
# Import necessary packages
import pandas as pd
import numpy as np
from datetime import datetime

# Begin coding here ...
# Use as many cells as you like


#Importing the CSV File
filename1 = "data/airbnb_price.csv"
airbnb_price = pd.read_csv(filename1)
print(airbnb_price)

#Importing the Excel File
filename2 = "data/airbnb_room_type.xlsx"
airbnb_room_type = pd.ExcelFile(filename2)
airbnb_room_type.sheet_names
df1=airbnb_room_type.parse("airbnb_room_type")
#print(df1)

#Importing the TSV File
filename3 = "data/airbnb_last_review.tsv"
airbnb_last_review = pd.read_csv(filename3, sep='\t')
#print(airbnb_last_review)

#Modifying the data type on last_review column from string to datetime
airbnb_last_review["last_review"].info()
airbnb_last_review["last_review_dates"] = pd.to_datetime(airbnb_last_review["last_review"])
#print(airbnb_last_review)


#What are the dates of the earliest and most recent reviews?

ordered_dates=list(airbnb_last_review["last_review_dates"].sort_values())
earliest_date=ordered_dates[0]
most_recent_date=ordered_dates[-1]
#print("The earliest date was " + str(earliest_date))
#print("The most recent date was " + str(most_recent_date))


#How many of the listings are private rooms?
#df1["room_type"].value_counts()
df1["room_type"]=df1["room_type"].astype("category")
df1["room_type"].cat.categories
update_categories = {"PRIVATE ROOM":"Private Room","Private room":"Private Room","private room":"Private Room"}
df1["room_type"]=df1["room_type"].replace(update_categories)
#print(df1["room_type"].dtype)

df2=df1["room_type"].value_counts()
private_rooms_count=df2[0]
print(df2)
print("The number of Private Rooms in the dataset is " + str(private_rooms_count))


#What is the average listing price? Round to the nearest two decimal places and save it into a variable
airbnb_price["price"]=airbnb_price["price"].str.replace("dollars","")
airbnb_price["price"]=airbnb_price["price"].astype("int")
#print(airbnb_price["price"].dtype)

avg_listing_price=round(airbnb_price["price"].mean(),2)
print(avg_listing_price)


#Combine the new variables into one DataFrame 
d={"first_reviewed":earliest_date,"last_reviewed":most_recent_date,"nb_private_rooms":private_rooms_count,"avg_price":avg_listing_price}
review_dates=pd.DataFrame(data=d, index=[0])
print(review_dates)
```
